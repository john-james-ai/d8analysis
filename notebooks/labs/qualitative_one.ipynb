## CategoricalOne

```python
s = list(df["Education"].values)
s = pd.Series(s)
s.index.sort_values()
s.nunique()
d = s.value_counts().to_dict()
d
pd.DataFrame.from_dict(data=d,orient='index')

c = s.value_counts().to_frame()
c.values
p = c/c.sum()*100
cp = c.cumsum() / c.sum() * 100
r = pd.concat([c,p,cp],axis=1)
r.columns = ["Count", "%", "Cum %"]
c
p
cp
r

```
