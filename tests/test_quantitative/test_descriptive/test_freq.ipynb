# Test Frequency Distribution
## Continuous

```python
from d8analysis.service.io import IOService
from d8analysis.container import D8AnalysisContainer
from d8analysis.quantitative.descriptive  import continuous, categorical
```

```python

container = D8AnalysisContainer()
container.init_resources()
container.wire(packages=["d8analysis"])
```

```python
FP = "data/Credit Score Classification Dataset.csv"
df = IOService.read(FP)
```

```python
fd = continuous.FrequencyDistribution(data=df, hue='Income', bins=4)
fd.compute()
fd.plot(title="Distribution of Income")
```

## Categorical


```python
fd = categorical.FrequencyDistribution(data=df, x='Education')
fd.compute()
fd.plot()
```
