```python
from d8analysis.quantitative.inferential.distribution.kstest import KSTest
from d8analysis.service.io import IOService
from scipy import stats
```

```python
from d8analysis.container import D8AnalysisContainer
container = D8AnalysisContainer()
container.init_resources()
container.wire(packages=["d8analysis.quantitative.inferential", "d8analysis.visual.seaborn"])
```

```python
FP = "data/Credit Score Classification Dataset.csv"
STATISTICS = ['ks', 'ad', 'cvm']
```

```python
df = IOService.read(FP, index_col=None)
a = df.loc[df['Gender'] == 'Male']['Income'].values
b = df.loc[df['Gender'] == 'Female']['Income'].values
```

```python

test = KSTest(a=df['Income'].values, b='norm')
test.run()
print(test.result)
test.result.plot()
```

```python
x = stats.norm.rvs(size=10000)
test = KSTest(a=x, b='norm')
test.run()
print(test.result)
test.result.plot()
```

```python
x = stats.norm.rvs(size=10000)

test = KSTest(a=x, b='norm')
test.run()
print(test.result)
test.result.plot()
```

```python
test = KSTest(a=a, b=b)
test.run()
print(test.result)
test.result.plot()
```
