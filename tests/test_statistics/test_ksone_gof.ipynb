# Kolmogorov-Smirnov Goodness of Fit Test

```python
from explorer.stats.goodness_of_fit.kstestone import KSTestOne
from explorer.service.io import IOService
import numpy as np
```

```python
FP = "data/Credit Score Classification Dataset.csv"
```

```python
df = IOService.read(FP, index_col=None)
test = KSTestOne()
test(data=df["Income"], reference_distribution='norm')
print(test.result)
test.result.plot()
```
