```python
from explorer.stats.distribution import RVSDistribution
from explorer.container import ExplorerContainer
from explorer.service.io import IOService
import matplotlib  as mpl
print(mpl.__version__)
```

```python
container = ExplorerContainer()
container.init_resources()
```

```python
FP = "data/Credit Score Classification Dataset.csv"
```

```python
df = IOService.read(FP, index_col=None)
```

```python
rvs = RVSDistribution()
rvs(data=df['Income'].values, distribution='uniform')
_ = rvs.histplot()
_ = rvs.histpdf()
_ = rvs.pdf()
_ = rvs.cdf()
_ = rvs.pdfcdf()
_ = rvs.ecdf()
```
