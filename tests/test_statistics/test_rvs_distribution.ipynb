# RVS Distribution  Test

```python
from explorer.stats.distribution import DISTRIBUTIONS, RVSDistribution
from explorer.service.io import IOService
```

```python
FP = "data/Credit Score Classification Dataset.csv"
```

```python
df = IOService.read(FP, index_col=None)
```

```python
rvs = RVSDistribution()
for dist in DISTRIBUTIONS.keys():
    rvs(data=df['Income'].values, distribution=dist)
    rvs.plot()

```
