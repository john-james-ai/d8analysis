# Shapiro-Wilk Test of Normality

```python
from d8analysis.stats.goodness_of_fit.shapiro import ShapiroWilkTest
from d8analysis.service.io import IOService
import numpy as np
```

```python
FP = "data/Credit Score Classification Dataset.csv"
```

```python
df = IOService.read(FP, index_col=None)
test = ShapiroWilkTest()
test(data=df["Age"])
print(test.result)
test.result.plot()
```
