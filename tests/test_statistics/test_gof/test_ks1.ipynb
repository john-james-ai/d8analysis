# Goodness of Fit Test

```python
from explorer.stats.goodness_of_fit.ksone import KSOneTest
from explorer.service.io import IOService
from scipy import stats
import numpy as np
from explorer.stats.distribution import DISTRIBUTIONS
```

```python
FP = "data/Credit Score Classification Dataset.csv"
STATISTICS = ['ks', 'ad', 'cvm']
```

```python
df = IOService.read(FP, index_col=None)

test = KSOneTest()
test(data=df["Income"], reference_distribution='normal')
print(test.result)
test.result.plot()
test.result.plotpdf()
test.result.plotcdf()
```

```python
test(data=df["Income"], reference_distribution='beta')
print(test.result)
test.result.plot()
test.result.plotpdf()
test.result.plotcdf()
```
