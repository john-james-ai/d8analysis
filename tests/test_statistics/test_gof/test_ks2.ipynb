# Goodness of Fit Test
## Two Sample

```python
from edation.stats.goodness_of_fit.kstwo import KSTwoTest
from edation.service.io import IOService
from scipy import stats
```

```python
FP = "data/Credit Score Classification Dataset.csv"
STATISTICS = ['ks', 'ad', 'cvm']
```

```python
df = IOService.read(FP, index_col=None)
female = df.loc[df['Gender'] == 'Female']['Income']
male = df.loc[df['Gender'] == 'Male']['Income']
female.name = 'female'
male.name = 'male'
female.head()
male.head()
```

```python

test = KSTwoTest()
test(sample1=female, sample2=male)
print(test.result)
test.result.plot()
test.result.plothist()
```
