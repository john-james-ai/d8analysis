```python
from edation.stats.centrality.ttest import TTest
from edation.service.io import IOService
from scipy import stats
```

```python
from edation.container import EdationContainer
container = EdationContainer()
container.init_resources()
```

```python
FP = "data/Credit Score Classification Dataset.csv"
STATISTICS = ['ks', 'ad', 'cvm']
```

```python
df = IOService.read(FP, index_col=None)
female = df.loc[df['Gender'] == 'Female']['Age']
male = df.loc[df['Gender'] == 'Male']['Age']
```

```python

test = TTest()
test(x=male, y=female)
print(test.result)
test.result.plot()
```
