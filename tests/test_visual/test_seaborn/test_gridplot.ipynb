```python
import warnings
import pandas as pd

import d8analysis as eda
warnings.filterwarnings(action='ignore', category=UserWarning)
```

```python
FP = "data/Credit Score Classification Dataset.csv"
df = pd.read_csv(FP, index_col=None)
df.info()
male = df.loc[df['Gender'] == 'Male']['Income'].values
female = df.loc[df['Gender'] == 'Female']['Income'].values
```

```python
p = eda.ECDFPlot(data=df, x='Income', title="Average Income\nProbability Density Function")
h = eda.Histogram(data=df, x='Income', title="Average Income\nHistogram")
plots = [p,h]
grid = eda.GridPlot()
for plot in plots:
    grid.add_plot(plot=plot)
grid.plot()
```
