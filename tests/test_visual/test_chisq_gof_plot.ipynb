# ChiSquare Goodness of Fit Test

```python
from d8analysis.quantitative.statistical.distribution.chisquare import ChiSquareResult
from d8analysis.visual.inferential import X2GOFTestPlot
from d8analysis.service.io import IOService
import numpy as np
```

```python
result = ChiSquareResult(
    test="Chi-Square Test of Goodness of Fit",
    value=5.085365853658536,
    pvalue=0.27865004658054116,
    alpha=0.05,
    dof=4,
    result=rf"$X^2$(4, N=164)=5.09, P=0.279 significant at 5%."
)
```

```python
p = X2GOFTestPlot(result=result)
p.plot()
```
