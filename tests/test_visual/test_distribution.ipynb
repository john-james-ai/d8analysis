# Distribution Plot Visualation Unit Tests

```python
import pandas as pd
from edation.visual.distribution import Histogram, KDEPlot, BoxPlot, ViolinPlot, ECDFPlot
```

```python
from edation.container import EdationContainer
container = EdationContainer()
container.init_resources()
```

## Data

```python
FP = "data/Credit Score Classification Dataset.csv"
df = pd.read_csv(FP, index_col=None)
```

```python
df.head()
```

## Histogram

```python
hist = Histogram()
hist(data=df,x="Age",hue='Gender', title="Test Histogram", stat="count", multiple='dodge')
```

## KDEPlot

```python
plot = KDEPlot()
plot(data=df,x="Income",hue='Gender', title="Test KDEPlot")
```

## Boxplot

```python
plot = BoxPlot()
plot(data=df,x="Income",y='Gender', title="Test Box Plot")
```

## Violin Plot

```python
plot = ViolinPlot()
plot(data=df,x="Income",y='Gender',  title="Test Violin Plot")
```

## ECDF Plot

```python
plot = ECDFPlot()
plot(data=df,x="Income",hue='Gender',  title="Test ECDF Plot")
```
