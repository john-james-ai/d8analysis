# Distribution Plot Visualation Unit Tests

```python
import pandas as pd
from d8analysis.visual.distribution import Histogram, PDF, BoxPlot, ViolinPlot, CDF
```

```python
from d8analysis.container import D8AnalysisContainer
container = D8AnalysisContainer()
container.init_resources()
```

## Data

```python
FP = "data/Credit Score Classification Dataset.csv"
df = pd.read_csv(FP, index_col=None)
```

```python
df.head()
```

## Histogram

```python
p = Histogram(data=df,x="Age",hue='Gender', title="Test Histogram", stat="count", multiple='dodge')
p.visualize()
```

## KDEPlot

```python
p = PDF(data=df,x="Income",hue='Gender', title="Test KDEPlot")
p.visualize()
```

## Boxplot

```python
p = BoxPlot(data=df,x="Income",y='Gender', title="Test Box Plot")
p.visualize()
```

## Violin Plot

```python
p = ViolinPlot(data=df,x="Income",y='Gender',  title="Test Violin Plot")
p.visualize()
```

## ECDF Plot

```python
p = CDF(data=df,x="Income",hue='Gender',  title="Test ECDF Plot")
p.visualize()
```
