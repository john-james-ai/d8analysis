# Association Plot Visualation Unit Tests

```python
import pandas as pd
from explorer.visual.association import ScatterPlot, LinePlot, PairPlot, JointPlot
```

```python
from explorer.container import ExplorerContainer
container = ExplorerContainer()
container.init_resources()
```

## Data

```python
FP = "data/Credit Score Classification Dataset.csv"
df = pd.read_csv(FP, index_col=None)
```

```python
df.head()
```

## Scatterplot

```python
plot = ScatterPlot()
plot(data=df,x="Age",y="Income", hue='Gender', title="Test Scatterplot")
```

## Lineplot

```python
plot = LinePlot()
plot(data=df,x="Age",y="Income", hue='Gender', title="Test Line Plot")
```

## Pair Plot

```python
plot = PairPlot()
plot(data=df, hue='Gender', title="Test Pair Plot")
```

## Joint Plot

```python
plot = JointPlot()
plot(data=df, x='Income', y="Age", hue='Gender', title="Test Joint Plot", height=6, ratio=6)
```
