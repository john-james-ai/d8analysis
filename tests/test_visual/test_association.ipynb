# Association Plot Visualation Unit Tests

```python
import pandas as pd
from edation.visual.association import ScatterPlot, LinePlot, PairPlot, JointPlot
```

```python
from edation.container import EdationContainer
container = EdationContainer()
container.init_resources()
```

## Data

```python
FP = "data/Credit Score Classification Dataset.csv"
df = pd.read_csv(FP, index_col=None)
```

```python
df.head()
```

## Scatterplot

```python
p = ScatterPlot(data=df,x="Age",y="Income", hue='Gender', title="Test Scatterplot")
p.visualize()
```

## Lineplot

```python
p = LinePlot(data=df,x="Age",y="Income", hue='Gender', title="Test Line Plot")
p.visualize()
```

## Pair Plot

```python
p = PairPlot(data=df, hue='Gender', title="Test Pair Plot")
p.visualize()
```

## Joint Plot

```python
p = JointPlot(data=df, x='Income', y="Age", hue='Gender', title="Test Joint Plot", height=6, ratio=6)
p.visualize()
```
