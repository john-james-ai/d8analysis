```python
import pandas as pd
from d8analysis.visual.distribution import Histogram, PDF, BoxPlot, ViolinPlot, CDF
from d8analysis.visual.visual import Visual
```

```python
from d8analysis.container import d8analysisContainer
container = d8analysisContainer()
container.init_resources()
```

```python
FP = "data/Credit Score Classification Dataset.csv"
df = pd.read_csv(FP, index_col=None)
```

```python
h = Histogram(data=df,x="Age",hue='Gender', title="Test Histogram", stat="count", multiple='dodge')
p = PDF(data=df,x="Income",hue='Gender', title="Test KDEPlot")
b = BoxPlot(data=df,x="Income",y='Gender', title="Test Box Plot")
v = ViolinPlot(data=df,x="Income",y='Gender',  title="Test Violin Plot")
c = CDF(data=df,x="Income",hue='Gender',  title="Test ECDF Plot")
plots = [h,p,b,v,c]
```

```python
v = Visual()
for p in plots:
    v.add_plot(plot=p)
v.visualize()
```
