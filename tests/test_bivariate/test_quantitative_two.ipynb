```python
import pandas as pd
import warnings
from explorer.analysis.bivariate import QualitativeTwo
from explorer.container import ExplorerContainer
from explorer.service.visual import LegendConfig, HistplotConfig, BarplotConfig, CountplotConfig, BoxplotConfig, PointplotConfig, KdeplotConfig
```

```python
container = ExplorerContainer()
container.init_resources()
container.wire(packages=["explorer"])
```

```python
DATAFILE = "data/Credit Score Classification Dataset.csv"
A = 'Education'
B1 = "Credit Rating"
B2 = "Income"
```

```python
df = pd.read_csv(DATAFILE)
a = QualitativeTwo(data=df)
```

## Describe

```python
a.describe(a=A,b=B1)
a.describe(a=A,b=B2)
```

## Frequency

```python
a.frequency(a=A,b=B1)
```

## Test Independence
### Two Categoricals

```python
print(a.test_independence(a=A,b=B1))
```

### Catgorical and Numeric

```python
print(a.test_independence(a=A,b=B2))
```
